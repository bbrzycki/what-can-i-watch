```python
import pandas as pd
df = pd.DataFrame(popular_movies['results'])
df.set_index('id', inplace=True)
df
```

```python
genre_dict = {
    28: "Action",
    12: "Adventure",
    16: "Animation",
    35: "Comedy",
    80: "Crime",
    99: "Documentary",
    18: "Drama",
    10751: "Family",
    14: "Fantasy",
    36: "History",
    27: "Horror",
    10402: "Music",
    9648: "Mystery",
    10749: "Romance",
    878: "Science Fiction",
    10770: "TV Movie",
    53: "Thriller",
    10752: "War",
    37: "Western"
}
```

```python
movies_df = df.reset_index().copy()
```

```python
# 1. Map IDs to genre names
movies_df['genre_names'] = movies_df['genre_ids'].apply(lambda id_list: [genre_dict.get(id) for id in id_list])

# 2. Use explode() to give each genre its own row, keeping the original id
exploded_genres = movies_df.explode('genre_names')

# 3. Use pandas.get_dummies to perform the one-hot encoding
genre_dummies = pd.get_dummies(exploded_genres['genre_names'], prefix='genre')

# 4. Concatenate the id column back and group by id
# This will create the final DataFrame where each row represents one movie
# and columns are the 0-1 genre indicators.
genre_indicator_df = pd.concat([exploded_genres['id'], genre_dummies], axis=1)
genre_indicator_df = genre_indicator_df.groupby('id').sum()

# Reset index to make 'id' a column again for merging
genre_indicator_df = genre_indicator_df.reset_index()

genre_indicator_df
```
